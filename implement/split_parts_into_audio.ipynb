```python
data_path = '.../emotion_data/'
audio_path = '.../emotion_data/audio_file/'
praise_path = '.../emotion_data/praise_part_data/'
not_praise_path = '.../emotion_data/not_praise_part_data/'
audio_fix_path = '.../emotion_data/audio_fix_error/'
```

```python
df_part_time = pd.ExcelFile(data_path + 'praise_speed_location.xlsx').parse('Sheet1')
```

```python
df_part_time.video.value_counts()
```

```python
def split_audio_part(startMin, startSec, endMin, endSec, audio_name , audio_part_name, part_path = praise_path, output_format = 'wav'):    
    # Time to miliseconds
    startTime = startMin*60*1000+startSec*1000
    endTime = endMin*60*1000+endSec*1000
    
    
    # Opening file and extracting segment
    song = pydub.AudioSegment.from_wav(audio_path + audio_name)
    extract = song[startTime:endTime]
    # Saving
    extract.export(part_path + audio_part_name, format= output_format)
```

```python
def split_praise_location_of_audio(audio_name):
    # location contain the praise audios of teacher    
    df = df_part_time[df_part_time.video == audio_name]
        
    for idx in df.index:
        if idx < 10:
            split_audio_part(df.loc[idx, 'startMin'], df.loc[idx, 'startSec'], df.loc[idx, 'endMin'], df.loc[idx, 'endSec'], audio_name, df.loc[idx, 'part_name'], part_path= praise_path)
        else:
            split_audio_part(df.loc[idx, 'startMin'], df.loc[idx, 'startSec'], df.loc[idx, 'endMin'], df.loc[idx, 'endSec'], audio_name, df.loc[idx, 'part_name'], part_path= praise_path)
```

```python
for audio in audio_files:
    split_praise_location_of_audio(audio)
```
